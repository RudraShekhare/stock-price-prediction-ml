```python
# Load Data 
df = pd.read_csv("../data/AAPL.csv")

# If no Date column, convert index to datetime
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])
    df.set_index("Date", inplace=True)
else:
    df.index = pd.to_datetime(df.index)
    df.index.name = "Date"

df = df.sort_index()
df.head()
```

```python


# Skip first 2 bad rows
df = pd.read_csv("../data/AAPL.csv", skiprows=2)

# Parse date
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

df = df.sort_index()
df = df[["Close", "Open", "High", "Low", "Volume"]]  # Drop extra columns if needed
df = df.apply(pd.to_numeric, errors='coerce')        # Convert all to numeric

df.head()
```
